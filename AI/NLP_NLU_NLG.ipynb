```python
from nltk.util import bigrams,trigrams,ngrams
string = 'Hello, best and most imortant thing in the worls cannot be seen, they must be felt'
quote_tokens = nltk.word_tokenize(string)
quote_tokens
```

```python
quotes_bigrams = list(nltk.bigrams(quote_tokens))
quotes_bigrams
```

```python
quotes_ngrams = list(nltk.ngrams(quote_tokens, 5))
quotes_ngrams
```

```python
from nltk import ne_chunk

ne_sent = 'The U.S President stays in Whitehouse'
```

```python
ne_tokens = word_tokenize(ne_sent)
ne_tokens
```

```python
ne_tags = nltk.pos_tag(ne_tokens)
ne_tags
```

```python
ne_ner = ne_chunk(ne_tags)
print(ne_ner)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
#create list of words

text = ("python python python python DataScience DataScience DataScience NLP NLU NLG NLP WorCLoud")
```

```python
text
```

```python
#create Word Cloud Object
wordcloud = WordCloud(width=420, height =200,margin=2, background_color='Red',colormap='Accent',mode='RGBA').generate(text)
```

```python
#Display the generated image

import matplotlib.pyplot as plt

# Display the generated image
plt.imshow(wordcloud, interpolation='bilinear')  # or 'quadratic'
plt.axis('off')
plt.margins(x=0, y=0)
plt.show()
```
